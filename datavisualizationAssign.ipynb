5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python

np.random.seed(42)



num_samples = 200
feature_1 = np.random.normal(0, 1, num_samples)
feature_2 = 2 * feature_1 + np.random.normal(0, 0.5, num_samples)  # Strongly correlated with feature_1
feature_3 = feature_1 + np.random.normal(0, 1, num_samples)        # Moderately correlated with feature_1
feature_4 = np.random.normal(0, 1, num_samples)                    # Not correlated
feature_5 = feature_2 - feature_3 + np.random.normal(0, 0.5, num_samples)  # Combination of correlated features


df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4,
    'Feature 5': feature_5
})


corr_matrix = df.corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set seed for reproducibility
np.random.seed(20)

# Generate sample data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix of Sales with Months as rows and Days as columns
heatmap_data = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='mean')

# Create the heatmap
fig = px.imshow(heatmap_data,
                labels=dict(x="Day", y="Month", color="Average Sales"),
                x=heatmap_data.columns,
                y=heatmap_data.index,
                title="Sales Variation Across Different Months and Days")

# Show the plot
fig.show()
```

Q4.Using the given x and y data, generate a 3D surface plot to visualize the function

```python


# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                 x=df['X'].values.reshape(100, 100),
                                 y=df['Y'].values.reshape(100, 100))])


fig.update_layout(title='3D Surface Plot of the Function',
                  scene=dict(
                      xaxis_title='X..',
                      yaxis_title='Y,,',
                      zaxis_title='Z//'
                 ))


fig.show()
```

 Q4.Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource

# Generate sample data
np.random.seed(42)
data_hist = np.random.randn(1000)  # 1000 normally distributed data points

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Convert the histogram data to a ColumnDataSource
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

# Create a Bokeh plot
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400,
           width=800)

# Add a quad glyph to represent the histogram
p.quad(top='top', bottom=0, left='left', right='right',
       source=source, fill_color="navy", line_color="white", alpha=0.6)

# Show the plot
output_file("histogram.html")  # Save the plot to an HTML file
show(p)
```

Q5.Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the grid and data to create a format suitable for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Define the color mapper
mapper = linear_cmap(field_name='value', palette="Viridis256", low=min(data_flat), high=max(data_flat))

# Create the Bokeh figure
p = figure(title="Heatmap of Random Data",
           x_axis_label='X',
           y_axis_label='Y',
           width=600,
           height=600,
           tools="",
           toolbar_location=None)

# Add rectangles (the cells of the heatmap)
p.rect(x="x", y="y", width=1/10, height=1/10, source=source,
       line_color=None, fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_file("heatmap.html")  # Save the plot to an HTML file
show(p)
```
